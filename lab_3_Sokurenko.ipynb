```python
# Bubble sort in Python

def bubbleSort(array):
    
  # loop to access each array element
  for i in range(len(array)):

    # loop to compare array elements
    for j in range(0, len(array) - i - 1):

      # compare two adjacent elements
      # change > to < to sort in descending order
      if array[j] > array[j + 1]:

        # swapping elements if elements
        # are not in the intended order
        temp = array[j]
        array[j] = array[j+1]
        array[j+1] = temp


data = [-7, 15, 3, 1, -23]

bubbleSort(data)

print('Sorted Array in Ascending Order:')
print(data)
```

```python
5. Дослідити реальний час виконання обох алгоритмів на Python і
побудувати графіки залежностей Tinsert(n) та Tbuble(n) для n =
[5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000],
згенерувавши відповідні цілочисельні масиви.
```

```python
import time
import random
import matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def generate_random_array(size):
    return [random.randint(0, 1000) for _ in range(size)]

def measure_time_sorting(algorithm, size):
    arr = generate_random_array(size)
    start_time = time.time()
    algorithm(arr)
    end_time = time.time()
    return end_time - start_time

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]
insertion_times = []
bubble_times = []

for size in sizes:
    insertion_time = measure_time_sorting(insertion_sort, size)
    bubble_time = measure_time_sorting(bubble_sort, size)
    insertion_times.append(insertion_time)
    bubble_times.append(bubble_time)

plt.plot(sizes, insertion_times, label='Insertion Sort')
plt.plot(sizes, bubble_times, label='Bubble Sort')
plt.xlabel('Array Size')
plt.ylabel('Time (s)')
plt.title('Sorting Algorithms Comparison')
plt.legend()
plt.show()
```

1. Що таке $O-нотація$  і чим вона відрізняється від $\Omega-нотації$?
$O-нотація$: Верхня межа зростання функції. Описує найгірший випадок. $f(n) = O(g(n))$ означає $f(n) \leq c \cdot g(n)$ для $n \geq n_0$.

$\Omega-нотація$: Нижня межа зростання функції. Описує найкращий випадок. $f(n) = \Omega(g(n))$ означає $f(n) \geq c \cdot g(n)$ для $n \geq n_0$.



2. Яку часову складність має алгоритм сортування методом вставки за найсприятливішої умови, коли масив відсортовано? Відповідь обгрунтувати. 


Алгоритм сортування методом вставки має найсприятливішу умову, коли масив вже відсортований. У цьому випадку, кожен елемент масиву порівнюється тільки з попереднім елементом, і оскільки масив відсортований, елемент завжди залишається на своєму місці. Тобто, внутрішній цикл, який відповідає за вставку, виконується тільки один раз для кожного елемента.

Часова складність у найсприятливішому випадку:
O(n)
де $n$ — кількість елементів у масиві.

Обґрунтування:
У відсортованому масиві кожен елемент порівнюється тільки один раз з попереднім, тому виконується $n-1$ порівнянь і пересувань, що дає лінійну часову складність.
