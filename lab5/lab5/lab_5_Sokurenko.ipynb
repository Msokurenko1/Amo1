Оцінити асимптотичну складність алгоритму лінійного пошуку у $О$-нотації.

```python
def linear_search(a_list, x):                
    i, length = 0, len(a_list)               
    while i < length and x != a_list[i]:    
        i += 1                                     
    return i+1 if i < length else -1       
    

search_item = int(input("Введіть шуканий елемент: "))

a = [32, 1, 3, 4, 2, 1]
linear_search(a, search_item)
```

* Оцінити асимптотичну складність алгоритму бінарного пошуку пошуку у $О$-нотації.

```python
def bin_search(a_list, x, left, right): 
    if left > right or len(a_list) == 0:     
        return -1
    middle = (left + right) // 2 
    if a_list[middle] == x:   
        return middle + 1          
    elif (a_list[middle] < x):                            
        return bin_search(a_list, x, middle + 1, right)    
    else: # a_list[middle] > x                             
        return bin_search(a_list, x, left, middle - 1)     


a = range(7,21)
search_item = int(input("Введіть шуканий елемент: "))
bin_search(a, search_item, 0, len(a) - 1)
```

 Час виконання бінарного пошуку зростає логарифмічно з розміром вхідних даних, і асимптотична складність становить $O(log(n))$

* за результатами попередніх завдань побудувати графіки залежності часу виконання процедури пошуку від розміру масиву. Приклад побудови графіків у середовищі `Ipython` 
* оцінити, який з двох алгоритмів є більш ефективним та в якому діапазоні розміру задачі.

```python
import matplotlib.pyplot as plt
import time

def linear_search(arr, target):
    start_time = time.time()
    for i in range(len(arr)):
        if arr[i] == target:
            break
    else:
        i = -1
    end_time = time.time()
    return end_time - start_time

def binary_search(arr, target):
    start_time = time.time()
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < target:
            low = mid + 1
        elif arr[mid] > target:
            high = mid - 1
        else:
            break
    else:
        mid = -1
    end_time = time.time()
    return end_time - start_time

def measure_search_time(search_func, array_sizes, target):
    times = []
    for size in array_sizes:
        arr = list(range(size))
        search_time = search_func(arr, target)
        times.append(search_time)
    return times

array_sizes = list(range(1000, 10001, 1000))  # Розміри масивів
target = array_sizes[-1] // 2  # Ціль для пошуку

linear_times = measure_search_time(linear_search, array_sizes, target)
binary_times = measure_search_time(binary_search, array_sizes, target)

plt.plot(array_sizes, linear_times, label='Linear Search')
plt.plot(array_sizes, binary_times, label='Binary Search')
plt.xlabel('Array Size')
plt.ylabel('Time (s)')
plt.title('Search Time vs Array Size')
plt.legend()
plt.grid(True)
plt.show()
```
