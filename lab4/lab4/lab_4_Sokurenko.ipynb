## Обчислення факторіалу.
Реалізувати процедуру обчислення факторіалу за допомогою цикла __for__. Оцінити асимптотичну складність алгоритму.

```python
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i

    return result
```

Асимптотична складність цього алгоритму = O(n).

Оцінити асимптотичну складність рекурсивного алгоритму обчислення факторіалу.

```python
def fac(n):
    
    # Обчислення факторіала через рекурсію
    
    if n == 0:
        return(1)
    return fac(n-1) * n
```

Асимптотична складність рекурсивного алгоритму обчислення факторіалу при n>0 буде O(n), а для т=0 асимптотична складність = O(1)

## Обчислення чисел Фібоначчі
Оцінити асимптотичну складність рекурсивного алгоритму обчислення $n$-го числа Фібоначчі.

```python
def fibonacci(n):
    
    #Рекурсивне обчислення n-го числа Фібоначчі
    
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2).
```

Асимптотична складність рекурсивного алгоритму обчислення n-го числа Фібоначчі= O(n), для n(0-2) і O(n<sup>2</sup>) для n>2


## Сортування злиттям
Оцінити асимптотичну складність алгоритму сортування злиттям, використовуючи основну теорему рекурсії.

```python
def merge(left, right):
    
    #Зливає два відсортованих масиви left і right у один
    
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(list):
    
    #Рекурсивна функція сортування з використанням попередньої функції
    
    if len(list) < 2:
        return list
    middle = len(list) // 2
    left = mergesort(list[:middle])
    right = mergesort(list[middle:])
    return merge(left, right)
```

Асимптотична складність алгоритму сортуванням злиттям =O(n<sup>2</sup>)
